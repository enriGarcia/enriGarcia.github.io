```python
from IPython.display import display, Math
import mpmath
from sympy import *

# Punto donde se evalúa la abscisa
abscisa = 1

x = symbols("x")

# Función de la curva
y = x ** 2 - 2
dy = diff(y, x)

# Pendiente e inclinación de la recta tangente
m = dy.subs(x, abscisa)
theta = N(mpmath.degrees(atan(m)))

b = y.subs(x, abscisa) - m * abscisa 

# Función de la línea tangente
t = m * x + b

display(Math(f' m = {latex(m)}'))
display(Math(f' \\theta = {latex(theta)} \\degree'))
p1 = plot(y, show=False)
p2 = plot(t, show=False)
p1.extend(p2)
p1.show()
```
